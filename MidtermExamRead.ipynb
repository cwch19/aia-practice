# 用迴圈把所有的資料合併在一起

```python
def merge_ds(datasets):
    result = pd.DataFrame()
    for dt in datasets.values():
        if result.empty:
            result = dt
        else:
            result = pd.merge(result, dt, on='date')
    
    return result
```
